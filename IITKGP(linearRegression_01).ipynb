**CREATING THE DATASET**

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-np.pi,np.pi, 100)
Y_function = 0.5*(x/2)+ np.sin(x)+np.random.random(x.shape)
plt.scatter(x, Y_function,color='black',marker='o')
plt.show()
```

**CHOOSING THE MODEL**

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()

```

**SPLITTING THE DATASET**

```python
random_indices=np.random.permutation(len(x))

x_train=x[random_indices[:70]]
y_train=Y_function[random_indices[:70]]

x_val=x[random_indices[70:85]]
y_val=Y_function[random_indices[70:85]]

x_test=x[random_indices[85:]]
y_test=Y_function[random_indices[85:]]
```

**FITTING THE MODEL**

```python
x_for_training=np.array(x_train.reshape(-1,1))
y_for_training=np.array(y_train.reshape(-1,1))
model.fit(x_for_training,y_for_training)
```

```python
plt.scatter(x_for_training,y_for_training,color='black',marker='o')
plt.plot(x_for_training,model.predict(x_for_training),color='blue',linewidth=1)
plt.show()
```

```python
mean_training_error=np.mean(y_for_training-np.array(model.predict(x_for_training).reshape(-1,1))**2)
print("MTE:",mean_training_error)
```

**VALIDATION SET**

```python
x_for_validation=np.array(x_val.reshape(-1,1))
y_for_validation=np.array(y_val.reshape(-1,1))
plt.scatter(x_for_validation,y_for_validation,color='black',marker='o')
plt.plot(x_for_validation,model.predict(x_for_validation),color='blue',linewidth=3)
plt.show()
```

**PREDICTING**

```python
model.predict(np.array(x_test.reshape(-1,1)))
```

**EVALUATION**

```python
mean_validation_error=np.mean((y_val - model.predict(x_val.reshape(-1,1)))**2)
mean_test_error=np.mean(y_test-model.predict(x_test.reshape(-1,1))**2)
print("MVE:",mean_validation_error)
print("MTE:",mean_test_error)
```
